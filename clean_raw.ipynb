```python
import pandas as pd
```

```python
# remove comments, emails, and
df = pd.read_csv('data/raw.csv')
df.drop("Email Address", axis=1, inplace=True)
df = df.drop(df.columns[[20, 23]], axis=1)
df.head()
```

```python
# compile course rankings and fix their individual column names
rankings = list(range(5, 19)) + [21, 22]
df["ranking"] = df[df.columns[rankings]].apply(
    lambda x: ','.join(x.astype(str)),
    axis=1
)
for i in range(len(rankings)):
    df.rename(columns = {df.columns[rankings[i]] :"ranking_" + str(i + 1)}, inplace = True)
df.head()
```

```python
# fix more column names
df.rename(
    columns = {
        df.columns[1]: "id",
        df.columns[2]: "major_status",
        df.columns[3]: "taken_courses",
        df.columns[4]: "grad_semester",
        df.columns[19]: "desired_amount_of_courses",
        df.columns[20]: "percieved_amount_of_courses_needed",
        }, 
    inplace = True
    )
df.head()
```

```python
# anonymize IDs
df["id"] = df.index
df.head()
```

```python
# save to csv
df.to_csv('data/clean.csv', index=False)
```

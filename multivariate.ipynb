```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = pd.read_csv('data/clean2.csv', index_col=0)
df.head()
```

```python
# linear regression on percieved need and desired amount of courses
x = df['desired_amount_of_courses']
y = df['percieved_amount_of_courses_needed']
model = LinearRegression()
model.fit(x.values.reshape(-1, 1), y.values.reshape(-1, 1))
xfit = np.linspace(0, 6) # just a list of 1000 values
yfit = model.predict(xfit[:, np.newaxis])
plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlabel('Desired amount of courses')
plt.ylabel('Percieved courses needed')
plt.title('Linear regression on desired courses and percieved need')
plt.show()
print("Model slope:", model.coef_[0])
print("Model intercept:", model.intercept_)
```
